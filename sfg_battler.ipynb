```python
import pyautogui
import cv2
import pydirectinput
import keyboard
import threading
import time
import os
from pynput import keyboard, mouse

mouse_controller = mouse.Controller()
clicking = False
moving = False

# Change the current working directory
os.chdir(r'C:/Users/Bobby/Desktop/My Python Stuff/sfg_forest')

def attack_click():
    """
    Searches for the "fight_button.png" image and clicks on it.
    """
    try:
        attackloc = pyautogui.locateOnScreen('fight_button.png', region=(650, 1290, 600, 130), confidence=0.9, grayscale=True)

        if attackloc:
            print("found attack image at", pyautogui.center(attackloc))
            attackx, attacky = pyautogui.center(attackloc)
            pydirectinput.moveTo(attackx, attacky)
            time.sleep(0.1)
            pydirectinput.moveTo(attackx - 1, attacky - 1)
            pydirectinput.click(attackx, attacky)

    except pyautogui.ImageNotFoundException:
        time.sleep(0.1)
        attackloc = False
        print("did not find attack image")
        pass
    
def loot_collector():
    """
    Searches for the "lootbox.png" image and clicks on it.
    """
    while True:
        try:
            lootloc = pyautogui.locateOnScreen('lootbox.png', region=(1650, 378, 70, 70), confidence=0.9, grayscale=True)
            if lootloc:
                print("found loot location", pyautogui.center(lootloc))
                lootx, looty = pyautogui.center(lootloc)
                pydirectinput.moveTo(lootx - 51, looty)  # Adjust offset as needed
                time.sleep(0.0001)
                pydirectinput.click(lootx - 51, looty)
                pydirectinput.moveTo(lootx - 50, looty)  # Adjust offset as needed

        except pyautogui.ImageNotFoundException:
            lootloc = False
            print("did not find loot image")
            pydirectinput.moveTo(1290, 1060)
            time.sleep(0.1)
            pydirectinput.moveTo(1290, 1059)
            pydirectinput.click(1290, 1060)
            break

def prioritize_images(image_list, confidence=0.9):
    """Finds images on the screen, prioritizing them based on their order in the list.

    Args:
        image_list (list): A list of image file paths to search for.
        confidence (float): The confidence level for image matching (0.0 - 1.0).

    Returns:
        tuple: The coordinates (left, top, width, height) of the first found image, or None if no images are found.
    """

    for image_path in image_list:
        try:
            location = pyautogui.locateOnScreen(image_path, confidence=confidence, grayscale=True)
            if location:
                return location
        except pyautogui.ImageNotFoundException:
            pass

    return None

# Example usage
image_list = ['sfg_450dmg.png', 'sfg_350dmg.png', 'sfg_300dmg.png', 'sfg_275dmg.png', 'sfg_225dmg.png', 'sfg_200dmg.png', 'sfg_150dmg.png', 'sfg_140dmg.png', 'sfg_100dmg.png', 'sfg_90dmg.png', 'sfg_50dmg.png', 'sfg_40dmg.png', 'sfg_20dmg.png']

def click_priority():
    location = prioritize_images(image_list, confidence=0.9)
    if location:
        print("Found monster priority at", pyautogui.center(location))
        loc_x, loc_y = pyautogui.center(location)
        pydirectinput.moveTo(loc_x, loc_y + 312)
        time.sleep(0.00001)
        pydirectinput.moveTo(loc_x, loc_y + 311)
        pydirectinput.click(loc_x, loc_y + 312)
    else:
        location = False
        print("moving onto attack search")
        pass

def on_press(key):
    global clicking
    try:
        if key.char == '8':
            clicking = True
            print("Clicking on!")
        elif key.char == '9':
            clicking = False
            print("Clicking off!")
    except:
        pass

def auto_clicker():
    while True:
        if clicking:
            prioritize_images(image_list, confidence=0.9)
            time.sleep(0.1)
            click_priority()
            time.sleep(2.6)
            attack_click()  # Call attack_click only once
            time.sleep(2.7)
            pydirectinput.moveTo(1684 - 51, 394)
            pydirectinput.moveTo(1684 - 51, 393)
            pydirectinput.click(1684 - 51, 394)
            time.sleep(2.6)
            loot_collector()  # Call loot_collector only once
        else:
            time.sleep(0.1)

def listen_for_keys():
    with keyboard.Listener(on_press=on_press) as listener:
        listener.join()

key_listener_thread = threading.Thread(target=listen_for_keys)
key_listener_thread.daemon = True
key_listener_thread.start()

# Start the auto_clicker
auto_clicker()
```
